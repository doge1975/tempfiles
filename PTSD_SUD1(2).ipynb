```python
def read_icd9_ICD10_Map(map_file):
    diag_dict = OrderedDict()
    f_diag = open(map_file, 'r')
    
    for line in f_diag:    
        ls = line.split('|')
        if ls[1] == 'Flags':
            continue
        if '.' in ls[0]:
            if ls[1] not in diag_dict:
                icd10_head='|'+ls[0].split('.')[0]+'|'
                diag_dict[ls[1]] = icd10_head
            
            else:
                icd10_head='|'+ls[0].split('.')[0]+'|'
                if not (icd10_head in diag_dict[ls[1]]):
                    diag_dict[ls[1]] =diag_dict[ls[1]]+icd10_head[1:]
                
            if ls[0] not in diag_dict:
                icd10_head='|'+ls[0].split('.')[0]+'|'
                diag_dict[ls[0]] =icd10_head
    return diag_dict
```
